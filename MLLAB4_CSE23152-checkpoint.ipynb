```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier

# Generate synthetic classification data
X, y = make_classification(n_samples=200, n_features=4, n_classes=2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Train kNN classifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
print("Train Confusion Matrix:\n", confusion_matrix(y_train, y_pred_train))
print("Test Confusion Matrix:\n", confusion_matrix(y_test, y_pred_test))
print("\nTrain Classification Report:\n", classification_report(y_train, y_pred_train))
print("\nTest Classification Report:\n", classification_report(y_test, y_pred_test))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

y_true = np.array([100, 150, 200, 250, 300])
y_pred = np.array([110, 140, 195, 260, 290])

mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
r2 = r2_score(y_true, y_pred)

print(f"MSE: {mse:.2f}, RMSE: {rmse:.2f}, MAPE: {mape:.2f}%, R2 Score: {r2:.2f}")
```

```python
import matplotlib.pyplot as plt
np.random.seed(42)
X_train = np.random.randint(1, 11, size=(20, 2))
y_train = np.array([0 if x[0] + x[1] < 12 else 1 for x in X_train])  
colors = ['blue' if label == 0 else 'red' for label in y_train]
plt.scatter(X_train[:, 0], X_train[:, 1], c=colors)
plt.xlabel("Feature X")
plt.ylabel("Feature Y")
plt.title("Training Data Scatter Plot")
plt.grid(True)
plt.show()
```

```python
x_vals = np.arange(0, 10.1, 0.1)
y_vals = np.arange(0, 10.1, 0.1)
test_points = np.array([[x, y] for x in x_vals for y in y_vals])
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
test_preds = knn.predict(test_points)

# Plot test classification
colors = ['blue' if label == 0 else 'red' for label in test_preds]
plt.scatter(test_points[:, 0], test_points[:, 1], c=colors, s=1)
plt.title("Test Data Classification (k=3)")
plt.xlabel("Feature X")
plt.ylabel("Feature Y")
plt.grid(True)
plt.show()
```

```python
for k in [1, 3, 5, 7, 9]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    test_preds = knn.predict(test_points)
    colors = ['blue' if label == 0 else 'red' for label in test_preds]
    plt.figure()
    plt.scatter(test_points[:, 0], test_points[:, 1], c=colors, s=1)
    plt.title(f"Test Classification with k={k}")
    plt.xlabel("Feature X")
    plt.ylabel("Feature Y")
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier


iris = load_iris()
X = iris.data[:, :2]  
y = iris.target


mask = y < 2
X = X[mask]
y = y[mask]

colors = ['blue' if label == 0 else 'red' for label in y]
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Training Data (Setosa vs Versicolor)")
plt.grid(True)
plt.show()

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))
test_points = np.c_[xx.ravel(), yy.ravel()]


for k in [1, 3, 5, 7, 9]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X, y)
    test_preds = knn.predict(test_points)
    colors = ['blue' if label == 0 else 'red' for label in test_preds]
    
    plt.figure(figsize=(8, 6))
    plt.scatter(test_points[:, 0], test_points[:, 1], c=colors, s=1, alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=['blue' if label == 0 else 'red' for label in y], edgecolors='k')
    plt.title(f"kNN Decision Boundary (k={k})")
    plt.xlabel("Sepal Length")
    plt.ylabel("Sepal Width")
    plt.grid(True)
    plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': list(range(1, 20))}
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid_search.fit(X_train, y_train)

print("Best k:", grid_search.best_params_['n_neighbors'])
print("Best score:", grid_search.best_score_)
```
